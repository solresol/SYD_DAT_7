```python
import spacy
```

```python
nlp = spacy.en.English()
```

```python
import codecs
```

```python
alice_file = codecs.open('../../data/alice-in-wonderland.txt', encoding='utf-8')
```

```python
alice = nlp(alice_file.read())
```

```python
alice
```

```python
alice.sents
```

```python
len(list(alice.sents))
```

```python
number_of_words = len(list(alice))
number_of_words
```

```python
number_of_distinct_words = len(set([t.orth_.lower() for t in alice if t.orth_.isalpha()]))
number_of_distinct_words
```

```python
lexical_diversity = 1.0 * number_of_distinct_words / number_of_words
lexical_diversity
```

```python
import collections
```

```python
verbs_in_alice = [(word.orth_, word.lemma_, word.rank) for word in alice
                  if word.pos_ == 'VERB'
                 if word.rank > 750000]
verbs_in_alice[:50]
```

```python
verbs_in_alice = [word.lemma_ for word in alice if word.pos_ == 'VERB']
c = collections.Counter()
c.update(verbs_in_alice)
```

```python
c.most_common()
```
