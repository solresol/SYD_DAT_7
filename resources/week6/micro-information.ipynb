```python
titanic.Embarked.unique()
```

```python
embark_coder = sklearn.preprocessing.LabelEncoder()
```

```python
titanic['port_as_a_number'] = embark_coder.fit_transform(titanic.Embarked.fillna('unknown'))
titanic.port_as_a_number
```

```python
embark_coder.classes_
```

```python
onehot = sklearn.preprocessing.OneHotEncoder(sparse=False)
```

```python
onehot.fit_transform(titanic[['port_as_a_number']]).shape
```

```python
onehot.fit_transform(titanic[['port_as_a_number']])
```

```python
hot_df = pandas.DataFrame(
    onehot.fit_transform(titanic[['port_as_a_number']]),
    columns = embark_coder.classes_
    )
hot_df
```

```python
titanic.merge(hot_df, left_index=True, right_index=True)
```

```python
sydtrains = pandas.read_csv('../../data/sydtrains.csv', index_col=1, thousands=',')
sydtrains
```

```python
all_lines = set(sydtrains.LINE)

all_lines = sydtrains.LINE.unique()

#Those two did basically the same thing... the first is a set, the other is an array
```

```python
import numpy
for line in all_lines:
    sydtrains[line + ' Line'] = numpy.where(sydtrains.LINE == line, 1, 0)
sydtrains
```

```python
len(all_lines)
```

```python
y = sydtrains.IN_24_HOURS
```

```python
X_columns = [x + " Line" for x in all_lines] + ['YEAR']
X_columns
```

```python
X = sydtrains[X_columns]
X
```

```python
import sklearn.grid_search
import sklearn.tree
```

```python
brute_force = sklearn.grid_search.GridSearchCV(
    sklearn.tree.DecisionTreeRegressor(),
    param_grid = {
        'max_depth': [1,2,3,4,5]
    }
)
```

```python
brute_force.fit(X, y)
```

```python
brute_force.best_params_
```

```python
brute_force.best_estimator_
```

```python
brute_force.best_score_
```

```python
brute_force.best_estimator_.predict([1] + ([0] * 19) + [2014])
```
