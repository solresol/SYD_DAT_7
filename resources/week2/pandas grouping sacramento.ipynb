```python
badness_score = pandas.DataFrame({
    'badness': complaints.INCIDENT_ZIP.value_counts()
    })
badness_score
```

```python
temp_df = badness_score.reset_index()
temp_df
```

```python
temp_df['clean_zip'] = temp_df['index'].astype(int)
temp_df
```

```python
temp_df['worstness'] = temp_df.badness.rank()
temp_df
```

```python
del temp_df['index']
temp_df
```

```python
temp_df.columns  = ['number_of_complaints', 'zipcode', 'horribleness_rank']
temp_df
```

```python
temp_df.set_index('zipcode', inplace=True)
temp_df
```

```python
where_not_to_buy = sacramento.merge(temp_df, left_on='zip', right_index=True)
where_not_to_buy
```

```python
# This would have worked if we hadn't done the set_index() above
sacramento.merge(temp_df, left_on='zip', right_on='zipcode')
```
